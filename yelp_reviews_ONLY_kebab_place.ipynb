```python
review_list =[]
```

```python
for review in total_reviews:
    parsed = review.text
    #print(parsed)
    review_list.append(parsed)
review_list    
```

```python
len(review_list)
```

```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
vader_results = []
vader_compound = []
vader_pos = []
vader_neu = []
vader_neg = []
```

```python
for r in review_list:

    # Run Vader Analysis on each Sample
    results = analyzer.polarity_scores(r)
    vader_results.append(r)
    
    compound = results["compound"]
    vader_compound.append(compound)
    
    pos = results["pos"]
    vader_pos.append(pos)
    
    neu = results["neu"]
    vader_neu.append(neu)
    
    neg = results["neg"]
    vader_neg.append(neg)

    # Print Samples and Analysis
    print(r)
    print("Compound Score:", compound)
    print("Positive Score:", pos)
    print("Neutral Score:", neu)
    print("Negative Score: ", neg)
    print()
```
