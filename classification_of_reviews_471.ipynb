```python
input_df = input_df.reset_index(drop=True)
input_df.head(30)
```

```python
# Set a default value
input_df['Class'] = 'neutral'
# Set Age_Group value for all row indexes which Age are greater than 40
input_df['Class'][input_df['Compound Score'] >= 0.05] = 'positive'
# Set Age_Group value for all row indexes which Age are greater than 18 and < 40
input_df['Class'][(input_df['Compound Score'] < 0.05) & (input_df['Compound Score'] > -0.05)] = 'neutral'
# Set Age_Group value for all row indexes which Age are less than 18
input_df['Class'][ -0.05 >= input_df['Compound Score']] = 'negative'
```
